```python
import pandas as pd
import numpy as np

# Haversine 거리 계산 함수 (단위: m)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371000  # 지구 반경 (m)
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)

    a = np.sin(dphi/2.0)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(dlambda/2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c  # meters

results = []
no_nearby_schools = []  # 근처 학교가 없는 폐교학교 기록

for _, close_row in close_school_1.iterrows():
    close_name = close_row["학교명"]
    close_students = close_row["학생수"]
    close_teachers = close_row["교사수"]
    close_ratio = close_row["수업교원 1인당 학생수"]
    close_lat, close_lon = close_row["위도"], close_row["경도"]

    # 각 폐교학교와 비폐교학교 거리 계산
    filtered_data["distance"] = filtered_data.apply(
        lambda row: haversine(close_lat, close_lon, row["위도"], row["경도"]), axis=1
    )

    # 500m 이내 학교만 선택
    nearby_schools = filtered_data[filtered_data["distance"] <= 500]

    if nearby_schools.empty:
        no_nearby_schools.append(close_name)
        continue

    # 가장 가까운 학교 선택
    target_row = nearby_schools.loc[nearby_schools["distance"].idxmin()]

    target_name = target_row["학교명"]
    student_before = target_row["학생수"]
    teacher_before = target_row["교사수"]
    ratio_before = target_row["수업교원 1인당 학생수"]

    student_after = student_before + close_students
    teacher_after = teacher_before + close_teachers
    ratio_after = student_after / teacher_after if teacher_after > 0 else np.nan

    results.append({
        "폐교학교명": close_name,
        "옮긴학교명": target_name,
        "학생수_전": student_before,
        "학생수_후": student_after,
        "교사수_전": teacher_before,
        "교사수_후": teacher_after,
        "수업교원 1인당 학생수_전": ratio_before,
        "수업교원 1인당 학생수_후": ratio_after
    })

final_df = pd.DataFrame(results)

# 최종 결과 확인
print("📌 최종 분배 결과:")
print(final_df.head())

# 근처 학교가 없는 폐교학교 확인
if no_nearby_schools:
    print("\n⚠️ 500m 이내에 학교가 없는 폐교학교:")
    for school in no_nearby_schools:
        print("-", school)
else:
    print("\n✅ 모든 폐교학교에 대해 500m 이내 학교가 존재합니다.")
```
