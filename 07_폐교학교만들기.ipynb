```python
merge_data['인구수'] = merge_data['인구수'].astype(int)
# 계산할 대상 컬럼들
facility_cols = ['지하철개수', '평생교육원기관수', '아동센터시설수', 
                 '노인복지관', '건강증진센터수', '체육시설수']

# 각 시설 컬럼별로 인구수 대비 비율 컬럼 생성
for col in facility_cols:
    merge_data[f'{col}_비율'] = merge_data[col] / merge_data['인구수'] * 100000

# 결과 확인
print(merge_data[['자치구', '인구수'] + [f'{col}_비율' for col in facility_cols]].head())
```

```python
merge_data.columns = merge_data.columns.str.strip()      # 앞뒤 공백 제거
merge_data.columns = merge_data.columns.str.replace("\n", "", regex=True)  # 줄바꿈 제거
```

```python
# 비율 컬럼들
ratio_cols = [
    '지하철개수_비율',
    '평생교육원기관수_비율',
    '아동센터시설수_비율',
    '노인복지관_비율',
    '건강증진센터수_비율',
    '체육시설수_비율'
]

# 각 비율 컬럼별 평균값 계산
mean_values = merge_data[ratio_cols].mean()

# 비교해서 '작다', '크다', '중간' 라벨링
for col in ratio_cols:
    mean_val = mean_values[col]
    merge_data[f'{col}_판정'] = merge_data[col].apply(
        lambda x: '작다' if x < mean_val else ('크다' if x > mean_val else '중간')
    )

# 결과 확인
print(merge_data[['자치구'] + ratio_cols + [f'{col}_판정' for col in ratio_cols]].head())

```

```python
pan = merge_data[['폐교명','자치구', '최다연령대', '지하철개수_비율_판정', '평생교육원기관수_비율_판정', '아동센터시설수_비율_판정', '노인복지관_비율_판정',
       '건강증진센터수_비율_판정', '체육시설수_비율_판정']]
pan
```

```python
# 연령대별 추천 시설 매핑
age_to_facility = {
    '20~29세': ['평생교육원기관'],
    '30~39세': ['아동센터시설'],
    '40~49세': ['건강증진센터', '체육시설'],
    '50~59세': ['건강증진센터', '체육시설'],
    '60~69세': ['노인복지관']
}

# 판정 컬럼과 시설 이름 매핑
facility_col_mapping = {
    '평생교육원기관수_비율_판정': '평생교육원기관',
    '아동센터시설수_비율_판정': '아동센터시설',
    '노인복지관_비율_판정': '노인복지관',
    '건강증진센터수_비율_판정': '건강증진센터',
    '체육시설수_비율_판정': '체육시설'
}

def 추천시설(row, df):
    age = row['최다연령대']
    recommended = []

    # 1️⃣ 최다연령대 맞춤 추천
    if age in age_to_facility:
        for target in age_to_facility[age]:
            for col, name in facility_col_mapping.items():
                if name == target and row[col] == '작다':
                    recommended.append(target)
    
    # 2️⃣ 추천 없으면 자치구 기반 추천
    if not recommended:
        gu = row['자치구']
        gu_df = df[df['자치구'] == gu]
        for col, name in facility_col_mapping.items():
            if row[col] == '작다' and name not in recommended:
                recommended.append(name)
    
    # 3️⃣ 그래도 추천 없으면 최다연령대 기준 시설 강제 추천
    if not recommended and age in age_to_facility:
        recommended = age_to_facility[age]
    
    return recommended if recommended else None

# 적용
pan['추천시설'] = pan.apply(lambda row: 추천시설(row, pan), axis=1)

# 확인
print(pan[['폐교명','최다연령대','추천시설']])
```
